```python
"""Преобразуем json-файл в датафрейм"""
import pandas as pd
import json

with open('hh_IT_month.json', 'r') as file:
    df1 = pd.json_normalize(json.load(file))
```

```python
df1
```

```python
df1.to_csv('hh_IT_month.csv')
```
